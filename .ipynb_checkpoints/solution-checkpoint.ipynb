```python
from numpy import array,zeros
import scipy as sc
from scipy import integrate
from run_kut4 import *
from printSoln import *
import pylab as pl
```

#### Part (a): Solve for the motion of the anharmonic oscillator

```python
#Define ODE as a system of equations by using z = x'
def F(t,x, w = 1.0):
    F = zeros(2)
    F[0] = x[1]
    F[1] = -w**2*x[0]**3
    return F

t = 0.0                  #Start of plot
tE = 20.0                #End of plot
x1 = array([1.0, 0.0])   #Initial conditions of x and z for amplitude A = 1
x10 = array([10.0, 0.0]) #Initial conditions of x and z for amplitude A = 10
h = 0.05                 #step size
freq = 40                #printout frequency

#Solve differential equation and print solutions for comparison and confirmation
T1,X1 = integrate(F,t,x1,tE,h)
#printSoln(T1,X1,freq)
T10,X10 = integrate(F,t,x10,tE,h)
#printSoln(T10,X10,freq)


#Plot the functions for A=1 and A=10 from t=0 to t=20
#A = 1
f, axs = pl.subplots(2,2,figsize=(15,3))
pl.subplot(121)
pl.plot(T1,X1[:,0],label='Position')
pl.xlabel('Time')
pl.title('Amplitude = 1')
pl.legend()


#A = 10
pl.subplot(122)
pl.plot(T10,X10[:,0],label='Position')
pl.xlabel('Time')
pl.title('Amplitude = 10')
pl.legend()
pl.show()
```

**Discussion**

In the above plots, we see that as we increase the amplitude A, the frequency of oscillation also increases.

#### Part (b): Make a *phase space plot* 

(i.e. $dx/dt$ vs $x$, instead of $x$ vs $t$)

```python
#Create phase space plots for the two different amplitudes
#A = 1
f, axs = pl.subplots(2,2,figsize=(15,3))
pl.subplot(121)
pl.axis('equal')
pl.plot(X1[:,0],X1[:,1])
pl.ylabel('Speed')
pl.xlabel('Position')
pl.title('Speed vs Position A = 1')


#A = 10
pl.subplot(122)
pl.axis('equal')
pl.plot(X10[:,0],X10[:,1])
pl.ylabel('Speed')
pl.xlabel('Position')
pl.title('Speed vs Position A = 10')
pl.show()
```

**Discussion**

These phase space plots are indeed reflective of oscillatory motion since we see that as the displacement reaches a maximum, the speed reaches 0, and as the speed reaches a maximum, the displacement reaches 0.
From these plots it is also apparent how the speed is directly proportional to the amplitude.

#### Part (c): Show how the behaviour in (a) and (b) differs (quantitatively) from a normal harmonic oscillator.  


```python
#(c) Harmonic Oscillator, same code as above for SHM, with the addition of the above plots for comparison
def HF(t,x, w = 1.0):
    F = zeros(2)
    F[0] = x[1]
    F[1] = -w**2*x[0]
    return F


#Solve Harmonic differential equation and print solutions for comparison and confirmation
HT1,HX1 = integrate(HF,t,x1,tE,h)
HT10,HX10 = integrate(HF,t,x10,tE,h)

#Print results for comparison
print('Compare Results: ')
printSoln(HT1,HX1,freq)
printSoln(T1,X1,freq)
print('\n')

#HARMONIC vs ANHARMONIC A = 1 (a)
f, axs = pl.subplots(2,2,figsize=(15,3))
pl.subplot(121)
pl.plot(HT1,HX1[:,0],label='Position')
pl.xlabel('Time')
pl.title('Harmonic Amplitude = 1')
pl.legend()


pl.subplot(122)
pl.plot(T1,X1[:,0],label='Position')
pl.xlabel('Time')
pl.title('Anharmonic Amplitude = 1')
pl.legend()
pl.show()

print('\r')
print('Comparing these two plots, we see that in the simple harmonic oscillator the period is shorter than that of the ')
print('\r' + ' anharmonic oscillator.' + '\n')

#HARMONIC vs ANHARMONIC A = 10 (a)
f, axs = pl.subplots(2,2,figsize=(15,3))
pl.subplot(121)
pl.plot(HT10,HX10[:,0],label='Position')
pl.xlabel('Time')
pl.title('Harmonic Amplitude = 10')
pl.legend()


pl.subplot(122)
pl.plot(T10,X10[:,0],label='Position')
pl.xlabel('Time')
pl.title('Anharmonic Amplitude = 10')
pl.legend()
pl.show()

print('\r')
print('Here we observe that in a harmonic oscillator, the speed and frequency do not change with the amplitude. However, in the an-')
print('\r' + 'harmonic oscillator, the speed increases as we change the amplitude from 1 to 10. The speed increases to the')
print('\r' + 'point where the frequency of the anharmonic oscillator clearly surpasses that of the harmonic one.' + '\n')

#HARMONIC vs ANHARMONIC A = 1 (b)
f, axs = pl.subplots(2,2,figsize=(15,3))
pl.subplot(121)
pl.axis('equal')
pl.plot(HX1[:,0],HX1[:,1])
pl.ylabel('Speed')
pl.xlabel('Position')
pl.title('Harmonic Speed vs Position A = 1')


pl.subplot(122)
pl.axis('equal')
pl.plot(X1[:,0],X1[:,1])
pl.ylabel('Speed')
pl.xlabel('Position')
pl.title('Anharmonic Speed vs Position A = 1')
pl.show()

print('\r')
print('In the above plots, the ideal relationship between position and speed in SHM is demonstrated by the circle, and we see ')
print('\r' + 'that it does not hold for anharmonic motion.' + '\n')
print('\r')

#HARMONIC vs ANHARMONIC A = 10 (b)
f, axs = pl.subplots(2,2,figsize=(15,3))
pl.subplot(121)
pl.axis('equal')
pl.plot(HX10[:,0],HX10[:,1])
pl.ylabel('Speed')
pl.xlabel('Position')
pl.title('Harmonic Speed vs Position A = 10')


pl.subplot(122)
pl.axis('equal')
pl.plot(X10[:,0],X10[:,1])
pl.ylabel('Speed')
pl.xlabel('Position')
pl.title('Anharmonic Speed vs Position A = 10')
pl.show()


print('\r')
print('Finally, comparing these last two plots, it is again shown how in harmonic oscillators, changes in the amplitude')
print('\r' + ' do not affect the speed of oscillation, unlike in anharmonic oscillators')
print('\r')
```
